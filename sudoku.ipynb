## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
"""
Function which checks if a move is valid with contrainsts.

Output:
True if it is valid to place the number at the given position, False otherwise.
"""

def is_valid_move(sudoku, row, col, num):
    # Check if the number is already in the row
    if num in sudoku[row]:
        return False
   
    # Check if the number is already in the column
    if num in [sudoku[i][col] for i in range(9)]:
        return False
   
    # Check if the number is already in the 3x3 grid
    grid_row = row // 3
    grid_col = col // 3
    for i in range(3):
        for j in range(3):
            if sudoku[grid_row*3 + i][grid_col*3 + j] == num:
                return False
   
    return True

"""
Function to get the minimum empty cell coordinates (MRV heuristics) in the sudoku puzzle and list of possible values

Output:
min_cell: A tuple (row, col) representing the cell with the minimum number of possible values
min_possible_values: A list of integers representing the possible values that can be placed in min_cell
"""

def min_empty_cells(sudoku):
   
    min_remaining_values = 10 #initialize to 10 as this is larger than max possible values a cell can have(9)
    min_cell = (None, None)
    min_possible_values = []
   
    for i in range(9):
        for j in range(9):
            if sudoku[i][j] == 0:
                # Count the number of possible values for this cell if valid move
                possible_values = [num for num in range(1,10) if is_valid_move(sudoku, i, j, num)]
                num_possible_values = len(possible_values)
               
                # Update the minimum cell if necessary
                if num_possible_values < min_remaining_values:
                    min_remaining_values = num_possible_values
                    min_cell = (i, j) #return the coordinates for the min_cel;
                    min_possible_values = possible_values #return a list of the possible values
                   
    return min_cell, min_possible_values


"""
Function to count the number of constraints that a given value would introduce if placed at a specific cell in the sudoku.
Called in solve_board to use LCV heuristics

Output:
An integer representing the number of constraints at that cell.
"""

def num_constraints(sudoku, row, col, value):
   
    num_constraints = 0
    # Check the row for constraints
    for i in range(9):
        if sudoku[row][i] == value:
            num_constraints += 1
           
    # Check the column for constraints
    for i in range(9):
        if sudoku[i][col] == value:
            num_constraints += 1
           
    # Check the 3x3 grid for constraints
    grid_row = row // 3
    grid_col = col // 3
    for i in range(3):
        for j in range(3):
            if sudoku[grid_row*3 + i][grid_col*3 + j] == value:
                num_constraints += 1
               
    return num_constraints



"""
Function to solve a Sudoku board using a recursive backtracking algorithm with MRV and LCV heuristics.

Output:
A boolean indicating whether the board was successfully solved or not.
"""

def solve_board(sudoku):
   
    min_cell, possible_values = min_empty_cells(sudoku)
    row, col = min_cell
   
    if min_cell[0] is None or min_cell[1] is None:
        return True
   
    # Order the possible values in ascending order of the number of constraints they introduce
    possible_values = sorted(possible_values, key=lambda z: num_constraints(sudoku, row, col, z))
   
    for num in possible_values:
        if is_valid_move(sudoku,row,col,num):
            sudoku[row][col] = num
           
            if solve_board(sudoku):
                return True
       
        #if not a valid move then backtrack and set it back to zero
        sudoku[row][col] = 0
   
    return False

"""
Helper functions to check for duplicate values in row_checker, col_checker and grid_checker.

Output:
Returns True if no duplicates and False otherwise, for each function respectively
"""

def row_checker(row, sudoku):

    # Extract the row as a list
    values = sudoku[row]

    values = [x for x in values if x != 0] # getting all non-zero values

    return len(values) == len(set(values))


def col_checker(col, sudoku):
   
    # Extract the column as a list
    values = [row[col] for row in sudoku]

    values = [y for y in values if y != 0] # getting all non-zero values

    return len(values) == len(set(values))


def grid_checker(sudoku):

    for row in range(0, 9, 3):
        for col in range(0,9,3):
            values = [sudoku[x][y] for x in range(row,row+3) for y in range(col, col+3) if sudoku[x][y] != 0]
         
            if len(values) != len(set(values)):
                return False
    return True


"""
Function to solve a sudoku puzzle.

Input:
sudoku: A 9x9 numpy array representing the puzzle. Empty cells are represented by 0.

Output:
A 9x9 numpy array representing the solved puzzle. If the puzzle cannot be solved, the output will be a 9x9 array of -1's.
"""

def sudoku_solver(sudoku):

    # Solve the puzzle using the solve_board() function
    solution = solve_board(sudoku)
   
    if solution is False:
        # Return a 9x9 array of -1's if the puzzle cannot be solved
        return np.full((9,9),-1)

    # Return the solved puzzle if it is valid and passes checks
    if all(row_checker(i, sudoku) and col_checker(i, sudoku) and grid_checker(sudoku) for i in range(9)):
        return sudoku
   
    # Otherwise, return a 9x9 array of -1's
    return np.full((9,9),-1)
```
